```python
# Question 8: Create a set of vowels from string "hello world"
print("\nQuestion 8: Create a set of vowels from string 'hello world'")
text = "hello world"
vowels = {'a', 'e', 'i', 'o', 'u'}
result = set()

for char in text:
    if char in vowels:
        result.add(char)

print(result)
```

```python
# Question 9: Remove duplicates using set
print("\nQuestion 9: Remove duplicates from list")
list1 = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
unique = set(list1)
print(unique)
```
